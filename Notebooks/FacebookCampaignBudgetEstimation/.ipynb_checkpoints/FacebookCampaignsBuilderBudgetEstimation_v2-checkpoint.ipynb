```python
from enum import Enum

from bson import BSON
from facebook_business.adobjects.adaccount import AdAccount
from numpy import array
from scipy.interpolate import interp2d

from Core.Tools.MongoRepository.MongoOperator import MongoOperator
from Core.Tools.MongoRepository.MongoRepositoryBase import MongoRepositoryBase
from Core.Web.BusinessOwnerRepository.BusinessOwnerRepository import BusinessOwnerRepository
from Core.Web.FacebookGraphAPI.GraphAPI.GraphAPISdkBase import GraphAPISdkBase
from Core.Web.FacebookGraphAPI.Tools import Tools
from Notebooks.FacebookCampaignBudgetEstimation.Startup import startup, logger
```

```python
class StructureStatusEnum(Enum):
    ACTIVE = 1
    REMOVED = 2
    DEPRECATED = 3
    PAUSED = 4
    ARCHIVED = 5
    COMPLETED = 6


class DataRepository(MongoRepositoryBase):

    def get_structures(self, level=None, ids=None, objective=None):
        self.set_database(self.config.structures_database_name)
        self.set_collection(level)

        query = {
            MongoOperator.AND.value: [
                {
                    'status': {
                        MongoOperator.IN.value: [StructureStatusEnum.ACTIVE.value,
                                                 StructureStatusEnum.PAUSED.value,
                                                 StructureStatusEnum.COMPLETED.value]
                    }
                }
            ]
        }
        if ids and level == 'adset':
            ids_condition = {
                'campaign_id': {
                    MongoOperator.IN.value: ids
                }
            }
            query[MongoOperator.AND.value].append(ids_condition)
        if objective:
            objective_condition = {
                'objective': {
                    MongoOperator.EQUALS.value: objective
                }
            }
            query[MongoOperator.AND.value].append(objective_condition)

        projection = {
            "_id": 0,
            "business_owner_facebook_id": 1,
            "account_id": 1,
            "campaign_id": 1,
            "adset_id": 1,
            "objective": 1,
            "details": 1
        }

        results = self.get(query=query, projection=projection)

        for result in results:
            if result['details']:
                result['details'] = BSON.decode(result['details'])

        return results

    def get_adsets_with_data(self, ids=None):
        self.set_database(self.config.insights_database_name)
        self.set_collection('adset_none_none')
        query = {
            MongoOperator.AND.value: [
                {
                    'adset_id': {
                        MongoOperator.IN.value: ids
                    }
                },
                {
                    'results': {
                        MongoOperator.NOTEQUAL.value: None
                    }
                },
                {
                    'amount_spent': {
                        MongoOperator.NOTEQUAL.value: None
                    }
                }
            ]
        }

        projection = {
            "_id": 0,
            "adset_id": 1
        }

        results = self.get(query=query, projection=projection)
        results = list(set([result['adset_id'] for result in results]))

        return results

    def get_insights(self, ids=None):
        self.set_database(self.config.insights_database_name)
        self.set_collection('adset_none_none')

        query = {
            MongoOperator.AND.value: [
                {
                    'adset_id': {
                        MongoOperator.IN.value: ids
                    }
                },
                {
                    'results': {
                        MongoOperator.NOTEQUAL.value: None
                    }
                },
                {
                    'amount_spent': {
                        MongoOperator.NOTEQUAL.value: None
                    }
                }
            ]
        }

        projection = {
            "_id": 0,
            "adset_id": 1,
            "results": 1,
            "amount_spent": 1,
            "date_start": 1
        }

        results = self.get(query=query, projection=projection)

        return results

def get_raw_data(repo=None, objective=None):
    # get campaigns
    campaigns = repo.get_structures(level='campaign', objective=objective)
    campaign_ids = [campaign['campaign_id'] for campaign in campaigns]

    # get adsets with data
    adsets = repo.get_structures(level='adset', ids=campaign_ids)
    adset_ids = [adset['adset_id'] for adset in adsets]
    active_adsets_ids = repo.get_adsets_with_data(ids=adset_ids)
    adsets = [adset for adset in adsets if adset['adset_id'] in active_adsets_ids]

    # append campaign objective to adsets with data
    adsets = append_objective_information(adsets=adsets, campaigns=campaigns)

    # append audience size to adsets with data
    adsets = append_audience_size_information(adsets=adsets)

    return adsets

def append_objective_information(adsets=None, campaigns=None):
    for adset in adsets:
        adset['objective'] = get_objective_by_campaign_id(adset['campaign_id'], campaigns)
    return adsets

def get_objective_by_campaign_id(campaign_id=None, campaigns=None):
    campaign = next(filter(lambda x: x['campaign_id'] == campaign_id, campaigns), None)
    if campaign:
        return campaign['objective']

def append_audience_size_information(adsets=None):
    for adset in adsets:
        adset['audience_size'] = get_audience_size(adset)
    return adsets

def get_audience_size(adset=None):
    bid = adset['business_owner_facebook_id']
    permanent_token = BusinessOwnerRepository(startup.session).get_permanent_token(bid)

    if permanent_token:
        _ = GraphAPISdkBase(startup.facebook_config, permanent_token)

        adset_details = adset['details']
        targeting_spec = adset_details.get('targeting', None)
        optimization_goal = adset_details.get('optimization_goal', None)

        ad_account_id = "act_" + adset['account_id']
        ad_account = AdAccount(fbid=ad_account_id)

        response = None

        if targeting_spec is not None and optimization_goal is not None:
            try:
                response = ad_account.get_delivery_estimate(fields=['estimate_mau'],
                                                            params={'targeting_spec': targeting_spec,
                                                                    'optimization_goal': optimization_goal})
                response = Tools.convert_to_json(response[0])
                response = response.get('estimate_mau')
            except Exception as e:
                pass

        return response

def aggregate_data(structure_info=None, insights=None):
    aggregated_data = []
    for structure in structure_info:
        adset_id = structure['adset_id']
        adset_insights = [insight for insight in insights
                          if insight['adset_id'] == adset_id]
        for adset_insight in adset_insights:
            adset_insight['objective'] = structure['objective']
            adset_insight['audience_size'] = structure['audience_size']
        aggregated_data += adset_insights
    return aggregated_data
```

```python
def budget_predictor(raw_data=None):
    x = [entry['audience_size'] for entry in raw_data if 'audience_size' in entry.keys()]
    x = array(x, dtype=float)

    y = [entry['results'] for entry in raw_data if 'audience_size' in entry.keys()]
    y = array(y, dtype=float)

    z = [entry['amount_spent'] for entry in raw_data if 'audience_size' in entry.keys()]
    z = array(z, dtype=float)

    f = interp2d(x, y, z, kind='cubic')
    return f

def estimate_budget(predictor=None, audience_size=None, results=None):
    return predictor(audience_size, results)
```

```python
objective = 'CONVERSIONS'
repo = DataRepository(config=startup.mongo_config, logger=logger)
data = get_raw_data(repo, objective)
```

```python
adset_ids = [entry['adset_id'] for entry in data]
raw_insights = repo.get_insights(ids=adset_ids)
aggregated_data = aggregate_data(structure_info=data, insights=raw_insights)
```

```python
import matplotlib.pyplot as plt
from copy import deepcopy
import numpy as np
from lmfit import Parameters, minimize, report_fit

def predict_budget(audience_size=None, results=None, cen_x=None, cen_y=None, sig_x=None, sig_y=None, offset=None):
    budget = gaussian2D(audience_size, results, cen_x, cen_y, sig_x, sig_y, offset)
    return budget

def gaussian2D(x, y, cen_x, cen_y, sig_x, sig_y, offset):
    return np.exp(-(((cen_x-x)/sig_x)**2 + ((cen_y-y)/sig_y)**2)/2.0) + offset

def residuals(p, x, y, z):
    height = p["height"].value
    cen_x = p["centroid_x"].value
    cen_y = p["centroid_y"].value
    sigma_x = p["sigma_x"].value
    sigma_y = p["sigma_y"].value
    offset = p["background"].value
    return (z - height*gaussian2D(x,y, cen_x, cen_y, sigma_x, sigma_y, offset))

raw_data = deepcopy(aggregated_data)

x = [entry['audience_size'] for entry in raw_data 
     if 'audience_size' in entry.keys() and entry.get('audience_size') > 0]
x = array(x, dtype=float)

y = [entry['results'] for entry in raw_data 
     if 'audience_size' in entry.keys() and entry.get('audience_size') > 0]
y = array(y, dtype=float)

z = [entry['amount_spent'] for entry in raw_data
     if 'audience_size' in entry.keys() and entry.get('audience_size') > 0]
z = array(z, dtype=float)

x_scaled = np.log10(x)
z_scaled = np.log10(z)

x_min = int(np.floor(min(x_scaled)))
x_max = int(np.ceil(max(x_scaled)))
y_min = int(np.floor(min(y)))
y_max = int(np.ceil(max(y)))

x_mesh, y_mesh = np.meshgrid(np.linspace(x_min, x_max, len(x)), np.linspace(y_min, y_max, len(y)))

initial = Parameters()
initial.add("height",value=1.)
initial.add("centroid_x",value=0.)
initial.add("centroid_y",value=0.)
initial.add("sigma_x",value=1.)
initial.add("sigma_y",value=3.)
initial.add("background",value=0.)

fit = minimize(residuals, initial, args=(x_mesh, y_mesh, z.ravel()))

#Â get predicted Gaussian params
cen_x = fit.params["centroid_x"].value
cen_y = fit.params["centroid_y"].value
sig_x = fit.params["sigma_x"].value
sig_y = fit.params["sigma_y"].value
offset = fit.params["background"].value
```

```python
# predict budget
z_pred = []
for i in range(len(x)):
    z_pred.append(gaussian2D(np.log10(x[i]), y[i], cen_x, cen_y, sig_x, sig_y, offset))
z_pred = array(z_pred)

z_pred = gaussian2D(x_mesh, y_mesh, cen_x, cen_y, sig_x, sig_y, offset)
```

```python
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_scaled, y, z_scaled, marker='o', s=40, c='r')
ax.plot_surface(x_mesh, y_mesh, z_pred)

ax.set_xlabel('Audience size')
ax.set_ylabel('Results')
ax.set_zlabel('Budget')

plt.show()
```
